```python
import patato as pat
import matplotlib.pyplot as plt
```

```python
rec_method = "Model Based"
clim = (0, 0.2)  # if "JAX" in rec_method else (-100, 420)
```

```python
pa1 = pat.PAData.from_hdf5("../Data/SKIN18/Scan_5.hdf5")
# OR SKIN18, 31
pa1.get_scan_name()
```

```python
pa2 = pat.PAData.from_hdf5("../Data/SKIN04/Scan_19.hdf5")
# OR SKIN04, 19 OR 01 , 7
pa2.get_scan_name()
```

```python
fig, axes = plt.subplots(2, 2, sharex=True, sharey=True, figsize=(8, 4))

# pa1.get_ultrasound().imshow(ax=axes[0, 0])
im = pa1.get_scan_reconstructions()[(rec_method, "0")].imshow(ax=axes[0, 0], clim=clim)

# im = pa1.get_scan_reconstructions()[(rec_method,'0')].imshow(ax=axes[0, 1], clim=clim)
im = pa1.get_scan_so2()[(rec_method, "0")].imshow(ax=axes[0, 1], clim=(0, 1))

# pa2.get_ultrasound().imshow(ax=axes[1, 0])
im = pa2.get_scan_reconstructions()[(rec_method, "0")].imshow(ax=axes[1, 0], clim=clim)

# im = pa2.get_scan_reconstructions()[(rec_method,'0')].imshow(ax=axes[1, 1], clim=clim)
im = pa2.get_scan_so2()[(rec_method, "0")].imshow(ax=axes[1, 1], clim=(0, 1))

plt.colorbar(im, ax=axes, label="PA Signal")
pa1.get_rois()["artery_", "0"].plot(ax=axes[0, 1])
pa2.get_rois()["artery_", "0"].plot(ax=axes[1, 1])
axes[0, 0].set_ylim([-0.01, 0.01])
# plt.tight_layout()

# axes[0, 0].set_title("Top (ITA = -25): THb ~ 4.7, sO2 ~ 38%", fontsize="small")
# axes[1, 0].set_title("Top (ITA = 65): THb ~ 9, sO2 ~ 65%", fontsize="small")

# SKIN18
axes[0, 0].set_title("Top (ITA = 25): THb ~ 14, sO2 ~ 77%", fontsize="small")
# SKIN04
axes[1, 0].set_title("Bottom (ITA = 34): THb ~ 7, sO2 ~ 67%", fontsize="small")
plt.show()
```

```python
fig, axes = plt.subplots(2, 2, sharex=True, sharey=True, figsize=(8, 4))

pa1.get_ultrasound().imshow(ax=axes[0, 0])
im = pa1.get_scan_reconstructions()[(rec_method, "0")].imshow(ax=axes[0, 1], clim=clim)

pa2.get_ultrasound().imshow(ax=axes[1, 0])
im = pa2.get_scan_reconstructions()[(rec_method, "0")].imshow(ax=axes[1, 1], clim=clim)

plt.colorbar(im, ax=axes, label="PA Signal")
pa1.get_rois()["skin_", "0"].plot(ax=axes[0, 1])
pa2.get_rois()["skin_", "0"].plot(ax=axes[1, 1])
axes[0, 0].set_ylim([-0.01, 0.01])
# plt.tight_layout()

axes[0, 0].set_title("Top (ITA = 25): THb ~ 14, sO2 ~ 77%", fontsize="small")
axes[1, 0].set_title("Bottom (ITA = 34): THb ~ 7, sO2 ~ 67%", fontsize="small")
plt.show()
```
