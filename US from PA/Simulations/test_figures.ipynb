```python
import numpy as np
import matplotlib.pyplot as plt

import matplotlib.patches as mpatches

from paiskintonetools import setup_matplotlib
from mpl_toolkits.axes_grid1.anchored_artists import AnchoredSizeBar
from mpl_toolkits.axes_grid1 import AxesGrid

setup_matplotlib(dpi=300)
```

```python
geometry = "linear"  # "linear" or "arc"
```

```python
datasets = [
    np.load(f"output_images/pa_{geometry}.npy") * 1e7,
    np.load(f"output_images/ruct_{geometry}.npy") * 10,
    np.load(f"output_images/mus_{geometry}.npy") * 1e8,
]
cmaps = ["gray", "bone", "magma"]
titles = ["PA$^{sim}$", "RUCT$^{sim}$", "mUS$^{sim}$"]
clims = [(-0.5, 2), (0.1, 0.7), (0.1, 1)]
# clims = [None, None, None]

if geometry == "linear":
    datasets = [datasets[0] / 1e6, datasets[1] / 10, datasets[2] / 1e6]
    datasets = [np.fliplr(x) for x in datasets]
```

```python
fig, axes = plt.subplots(1, 3, figsize=(6, 1.5))

axes[0].plot([0, 0], [-0.008, -0.008 + 0.0065], c="C1")
axes[0].plot([0, 0], [-0.0015, -0.0015 + 0.0065], c="C1")

axes[2].plot([0, 0], [-0.0015, -0.0015 + 0.0065], c="C1")

for ax, data, cmap, title, clim in zip(axes, datasets, cmaps, titles, clims):
    im = ax.imshow(data, cmap=cmap, origin="lower", extent=(-0.02, 0.02) * 2, clim=clim)
    ax.set_title(title)
    plt.colorbar(im, ax=ax)
for ax in axes:
    ax.yaxis.set_tick_params(labelleft=False)
    ax.xaxis.set_tick_params(labelbottom=False)


asb = AnchoredSizeBar(
    axes[0].transData,
    0.01,
    "1 cm",
    loc="lower right",
    pad=0.1,
    borderpad=0.5,
    sep=5,
    frameon=False,
    color="white",
)
axes[0].add_artist(asb)

plt.show()
```
